```python
import re
import matplotlib.pyplot as plt
import ast

# Function to read metrics from text file using regex
def plot_metrics(filename):
    metrics = {}
    with open(filename, 'r') as file:
        content = file.read()
        # Extract all occurrences of loss using regex
        loss = [ast.literal_eval(item) for item in re.findall(r'\nloss:(.*?)\n', content)]
        mse = [ast.literal_eval(item) for item in re.findall(r'\nmse:(.*?)\n', content)]
        rmse = [ast.literal_eval(item) for item in re.findall(r'\nrmse:(.*?)\n', content)]
        snr = [ast.literal_eval(item) for item in re.findall(r'\nsnr:(.*?)\n', content)]
        psnr = [ast.literal_eval(item) for item in re.findall(r'\npsnr:(.*?)\n', content)]
        val_loss = [ast.literal_eval(item) for item in re.findall(r'\]\nval_loss:(.*?)\n', content)]
        val_mse = [ast.literal_eval(item) for item in re.findall(r'\nval_mse:(.*?)\n', content)]
        val_rmse = [ast.literal_eval(item) for item in re.findall(r'\nval_rmse:(.*?)\n', content)]
        val_snr = [ast.literal_eval(item) for item in re.findall(r'\nval_snr:(.*?)\n', content)]
        val_psnr = [ast.literal_eval(item) for item in re.findall(r'\nval_psnr:(.*?)\n', content)]
        epoch_time = [ast.literal_eval(item) for item in re.findall(r'\nEpoch Time:(.*?)\n', content)]
        metrics['loss'] = loss
        metrics['mse'] = mse
        metrics['rmse'] = rmse
        metrics['snr'] = snr
        metrics['psnr'] = psnr
        metrics['val_loss'] = val_loss
        metrics['val_mse'] = val_mse
        metrics['val_rmse'] = val_rmse
        metrics['val_snr'] = val_snr
        metrics['val_psnr'] = val_psnr
        metrics['epoch_time'] = epoch_time

    return metrics
```

```python
client1_metrics = plot_metrics('Results/classical_vs_federated/fl4_prox/client1_metrics.txt')
print(client1_metrics)
for client1_metrics_key, client1_metrics_list in client1_metrics.items():
    fig, ax = plt.subplots()
    
    for i, client1_metrics_sublist in enumerate(client1_metrics_list):
        ax.plot(client1_metrics_sublist, label=f'Round {i+1}')

    ax.set_xlabel('Epoch')
    ax.set_ylabel(client1_metrics_key)
    ax.set_title(f'{client1_metrics_key} vs. Epoch')
    ax.legend()
    
    #plt.show()
    plt.savefig(f'Results/classical_vs_federated/fl4_prox/plots/client1_{client1_metrics_key}vsEpoch.png')
```
